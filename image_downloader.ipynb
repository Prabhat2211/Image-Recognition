```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import os
import time

# Function to create a directory if it doesn't exist
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

query = input('What are you looking for? ')
n_images = int(input('How many images do you want? '))

url = 'https://www.google.com/search?q='+query+'&tbm=isch' 
driver = webdriver.Safari()


driver.get(url)


scroll_count = 5  
for i in range(scroll_count):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1) 

page_source = driver.page_source


driver.quit()


soup = BeautifulSoup(page_source, 'html.parser')


image_directory = 'images/zomato'
create_directory(image_directory)

# Find all image elements on the page
image_elements = soup.find_all('img')

# Download and save a specific number of images (e.g., 5 out of 20)
download_limit = n_images
downloaded_count = 0
for img_element in image_elements:
    if downloaded_count >= download_limit:
        break
    image_url = img_element.get('src')
    if image_url and 'http' in image_url:  # Ensure it's a valid image URL
        image_content = requests.get(image_url).content
        image_name = os.path.join(image_directory, f"image_{319+downloaded_count}.jpg")
        with open(image_name, 'wb') as image_file:
            image_file.write(image_content)
            print(f"Image '{image_name}' downloaded.")
        downloaded_count += 1
```
